Este problema se resuelve utilizando el Teorema del Sandwich.
Sabemos que para cualquier número real \( n \), la función seno está acotada entre -1 y 1:

$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

El Teorema del Sandwich dice que si una sucesión \( a_n \) está atrapada entre dos sucesiones \( b_n \) y \( c_n \), y ambas convergen al mismo límite, entonces \( a_n \) también converge a ese límite.

En este caso, observamos que:

$$
\lim_{n \to \infty} -\frac{1}{n} = 0 \quad \text{y} \quad \lim_{n \to \infty} \frac{1}{n} = 0.
$$

$$
\text{Para } n = 1:
$$

$$
\frac{\sin(1)}{1} = \sin(1) \approx 0.8415
$$

$$
\text{Para } n = 2:
$$

$$
\frac{\sin(2)}{2} = \frac{0.9093}{2} = 0.4547
$$

$$
\text{Para } n = 3:
$$

$$
\frac{\sin(3)}{3} = \frac{0.1411}{3} = 0.0470
$$

El límite es 0 porque el numerador (
sin
⁡
(
𝑛
)
sin(n)) está acotado entre -1 y 1, mientras que el denominador (
n) crece indefinidamente, haciendo que la fracción se haga cada vez más pequeña y mas cercana a cero, por ende:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 100)
sin_n_sobre_n = np.sin(n) / n
uno_sobre_n = 1 / n
menos_uno_sobre_n = -1 / n


plt.figure(figsize=(8, 6))
plt.plot(n, sin_n_sobre_n, linestyle = '-')
plt.plot(n, uno_sobre_n, 'g--')
plt.plot(n, menos_uno_sobre_n, 'b--')

plt.axhline(y=0, color='black', linestyle='--')

plt.title('Teorema del Sandwich')
plt.grid()

plt.show()
```

```python
import numpy as np
import sympy as sp

from IPython.display import display, Math

def geometric_sum(a, r):
    if abs(r) < 1:
        return a / (1 - r)
    else:
        return None

results = {}

a, r = 1, 1/3
results['(a)'] = geometric_sum(a, r)

a, r = 4, 0.21
results['(b)'] = geometric_sum(a, r)

a, r = 2/5, 1/5
results['(c)'] = geometric_sum(a, r)

a, r = 2, 1/7
results['(d)'] = geometric_sum(a, r)

a, r = (7 * (1/3) * (2/5)), (1/3) * (2/5)
results['(e)'] = geometric_sum(a, r)

a, r = (2/3)**3, 2/3
results['(f)'] = geometric_sum(a, r)

a, r = 1, 0.99999
results['(h)'] = geometric_sum(a, r)

a, r = sp.pi/4, sp.pi/4
results['(l)'] = geometric_sum(a, r)


for key, value in results.items():
    if value is not None:
        display(Math(rf"{key}: S = {sp.latex(value)}"))
print("El resto de las series no convergen, por lo que no tienen una suma ")
```

   $$

$$
S = \frac{a}{1 - r}
$$

(a = 7.5\) (primer término después de la caída).

$$
S = \frac{7.5}{1 - 0.75} = \frac{7.5}{0.25} = 30
$$

(r = 0.25\)

$$

$$

$$

### **Cálculo de ( n ) para que ( H_n > S )**

La serie armónica está definida como:

$$
H_n = \sum_{k=1}^{n} \frac{1}{k}
$$
Sabemos que la serie armónica se **aproxima** asintóticamente por la expresión:

$$
H_n \approx \ln(n) + \gamma
$$

donde ( gamma ≈ 0.5772 ) es la **constante de Euler-Mascheroni**.

Queremos encontrar el menor ( n ) tal que:

$$
H_n > S
$$

Lo resolvemos despejando \( n \):

$$
\ln(n) + \gamma > S
$$

$$
\ln(n) > S - \gamma
$$

$$
n > e^{S - \gamma}
$$

#### **Cálculo para \( S = 5 \)**
Sustituyendo \( S = 5 \):

$$
n > e^{5 - 0.5772}
$$

$$
n > e^{4.4228}
$$

$$
n \approx 83
$$

Por lo tanto, **el menor valor de ( n ) que cumple ( H_n > 5 ) es ( 83 )**.

---

#### **Cálculo para \( S = 10 \)**
Sustituyendo \( S = 10 \):

$$
n > e^{10 - 0.5772}
$$

$$
n > e^{9.4228}
$$

$$
n \approx 12367
$$

Por lo tanto, **el menor valor de ( n ) que cumple ( H_n > 10 ) es ( 12367 )**.

---

### **Verificación Numérica**
Los valores aproximados de la serie armónica confirman:


$$H_{83} ≈ 5.008$$



$$H_{12367} ≈ 10.0001$$


Esto coincide con nuestras aproximaciones teóricas.
