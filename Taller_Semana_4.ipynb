```python
import matplotlib.pyplot as plt

def fibonacci(n):
    fib_series = [0, 1]
    for i in range(2, n):
        next_num = fib_series[-1] + fib_series[-2]
        fib_series.append(next_num)
    return fib_series[0:n]


n = int(input("Introduce el número de términos de la sucesión de Fibonacci que deseas generar: "))
fib_series = fibonacci(n)

plt.plot(fib_series, marker='o', linestyle='-', color='k')
plt.title(f"Serie de Fibonacci con {n} términos")
plt.xlabel("Índice")
plt.ylabel("Número de Fibonacci")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generacion_numeros():
    return np.random.randint(18, 35, 365)

def media(datos):
    return np.mean(datos)

def mediana(datos):
    return np.median(datos)

def desviacion_estandar(datos):
    return np.std(datos)

def graficar_temperatura_mensual(datos):
    meses = [i // 30 for i in range(365)]
    temperatura_mensual = [np.mean([datos[i] for i in range(365) if meses[i] == mes]) for mes in range(12)]

    plt.plot(range(1, 13), temperatura_mensual, marker='o', linestyle='-', color='r', label='Media Mensual')
    plt.xlabel('Mes')
    plt.ylabel('Temperatura (°C)')
    plt.title('Temperatura Media Mensual')
    plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
    plt.grid()
    plt.legend()
    plt.show()


temperaturas = generacion_numeros()
print(f"Media: {media(temperaturas):.2f}°C")
print(f"Mediana: {mediana(temperaturas):.2f}°C")
print(f"Desviación Estándar: {desviacion_estandar(temperaturas):.2f}°C")
graficar_temperatura_mensual(temperaturas)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 500)

cos = np.cos(x)
sen = np.sin(x)
tan = np.tan(x)

plt.figure(figsize=(10,6))
plt.plot(x,cos, label='coseno', color='b')
plt.plot(x,sen, label='seno', color='r')
plt.plot(x,tan, label='tangente', color='g')

plt.ylim(-5, 5)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico seno, coseno y tangente')

plt.grid(linestyle='--')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = float(input("Ingrese la coordenada x del primer punto: "))
y1 = float(input("Ingrese la coordenada y del primer punto: "))
x2 = float(input("Ingrese la coordenada x del segundo punto: "))
y2 = float(input("Ingrese la coordenada y del segundo punto: "))

m = (y2 - y1) / (x2 - x1)

print("La pendiente de la recta es:", m)

x = np.linspace(x1 - 2, x2 + 2, 100)
y = m * (x - x1) + y1

plt.plot(x, y, label=f'y = {m:.2f}x + {y1 - m * x1:.2f}', color='blue')
plt.plot(x1, y1, '.',color='k', ms=10, label='Punto 1')
plt.plot(x2, y2, '.',color='k', ms=10, label='Punto 2')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfica de la recta')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v0 = 20
g = 9.8

angulos=[]
for i in range(5):
  angulo = int(input("Escribe el angulo de lanzamiento: "))
  angulos.append(angulo)

for i in angulos:
    theta = np.radians(i)
    t_vuelo = (2 * v0 * np.sin(theta)) / g
    t = np.linspace(0, t_vuelo, 100)

    x = v0 * np.cos(theta) * t
    y = v0 * np.sin(theta) * t - 0.5 * g * t**2

    plt.plot(x, y, label=f'{i}°')

plt.title('Movimiento parabólico para diferentes ángulos')
plt.xlabel('Distancia (m)')
plt.ylabel('Altura (m)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

def sierpinski(n):
  x = [0.5]
  y = [0.0]

  for i in range(n):
    k = random.randint(0, 2)
    if k == 0:
      x.append(0.5 * x[-1])
      y.append(0.5 * y[-1])
    elif k == 1:
      x.append(0.5 * x[-1] + 0.25)
      y.append(0.5 * y[-1] + 0.5 * (3**0.5)/2)
    else:
      x.append(0.5 * x[-1] + 0.5)
      y.append(0.5 * y[-1])

  return x, y

n = 150000
x, y = sierpinski(n)

plt.figure(figsize=(6.5,6))
plt.scatter(x, y,s= 0.1, marker='o', color='grey')
plt.title("Fractal de Sierpinski")
plt.show()
```
